```python
import json
from os import listdir, path



def winPath(path):
    p = path.replace('\\','/')
    p = p.replace('//','/')
    return p

def loadEmojis(folder):
    files = [f for f in listdir(folder) if path.isfile(path.join(folder, f))]
    emojis = {}
    names = []
    for file in files:
        name, ext = path.splitext(file)
        if ext == '.json':
            raw = json.load(open(folder + file, 'r'))
            name = raw.keys()[0]
            category = raw[name]
            emojis[name] = category[category.keys()[0]]
    return emojis
        

folder = winPath(r'D:\edu\phd\Computational Interaction\EmojiCodeSheet\json\string\\')

emojis = loadEmojis(folder)            
    
```
